### 0. an end-to-end scikitlearn workflow

## 1. Get the data ready

```python
import pandas as pd
heart_disease = pd.read_csv("heart-disease.csv")
```

```python
heart_disease
```

```python
# Create X (feature matrix)
x = heart_disease.drop("target", axis=1)

# Create Y (labels)
y = heart_disease["target"]
```

```python
# 2. Choose the right model
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier()
clf.get_params()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3)
```

```python
clf.fit(x_train, y_train)
```

```python
# Make Prediction
import numpy as np
# y_label = clf.predict(np.array([0,2,3,4]))
```

```python
y_prdi = clf.predict(x_test)
y_prdi
```

```python
y_test
```

```python
# 4. evaluate the model 
clf.score(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print (classification_report(y_test, y_prdi))
```

```python
confusion_matrix(y_test, y_prdi)
```

```python
accuracy_score(y_test, y_prdi)
```

```python
# 5. Imporve model 
np.random.seed(42)
for i in range(10,100,10):
    print(f"Tring  model with {i} estimator..")
    clf = RandomForestClassifier(n_estimators=i).fit(x_train, y_train)
    print(f"model accurarcy on test set: {clf.score(x_test, y_test) * 100:.2f}%")
    
```

```python
# 6. save a model and load
import pickle
pickle.dump(clf, open("random_forest_model.pkl", "wb"))
```

```python
load_model = pickle.load(open("random_forest_model.pkl","rb"))
kk = load_model.score(x_test, y_test)
kk
```
